```python
import customtkinter as ctk
from tkinter import messagebox as mb
import pandas as pd

# Standardizing the alias to ctk
ctk.set_appearance_mode("dark")
ctk.set_default_color_theme("dark-blue")

app = ctk.CTk()
app.title("Employee Portal")
app.geometry("400x500") # Increased height slightly for better layout

# Load the data once at the start
try:
    df = pd.read_csv("data.csv")
except FileNotFoundError:
    # Creating dummy data for testing if file is missing
    df = pd.DataFrame({"Name": ["Alice", "Bob"], "Department": ["hr", "tech"]})

def check_employee():
    name = nameentry.get().strip()
    dept = dptentry.get().strip()
    
    # Matching both name and department (case-insensitive)
    match = df[(df["Name"].str.lower() == name.lower()) & 
               (df["Department"].str.lower() == dept.lower())]
    
    
    
    
    
    
    
    resultdisplay.configure(state="normal")
    resultdisplay.delete("1.0", "end")
    
    if not match.empty:
        resultdisplay.insert("end", f"Success: Employee {name} found in {dept} department.\n")
        resultdisplay.insert("end", "_" * 30 + "\n")
    else:
        mb.showwarning("Not Found", f"Employee {name} not found in {dept}.")
        resultdisplay.insert("end", f"Error: No match for {name} in {dept}.\n")
    
    resultdisplay.configure(state="disabled")

# UI Elements
title = ctk.CTkLabel(app, text="Employee Checker", font=("Arial", 20, "bold"))
title.pack(pady=20)

nameentry = ctk.CTkEntry(app, placeholder_text="Enter employee name", width=250)
nameentry.pack(pady=10)

dptentry = ctk.CTkEntry(app, placeholder_text="Enter department name", width=250)
dptentry.pack(pady=10)

# Fixed the command argument here
verifybutton = ctk.CTkButton(app, text="Check Database", command=check_employee)
verifybutton.pack(pady=20)

resultdisplay = ctk.CTkTextbox(app, width=350, height=150, state="disabled")
resultdisplay.pack(pady=10)

app.mainloop()
```
