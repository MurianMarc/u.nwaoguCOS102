```python
class House:
    def __init__(self, house_type, color, room):
        self.house_type = house_type
        self.color = color
        self.room = room
        
    def introduce(self):
        print(f"My {self.house_type} house is colored {self.color} and has {self.room} rooms.")
    def paint(self):
        while True:
            new_color = input("What color do you want to paint your house? ")
            choice = input(f"Do you want to paint your house {new_color}? (yes/no) ")
            
            if choice.lower() == "yes":
                self.color = new_color
                
                print(f"My {self.room}-room {self.house_type} house is now painted {self.color}")
                break
            
            elif choice.lower() == "no":
                print("Okay, your house will remain the same color.")
                break
            
            else:
                print("Invalid input. Please enter 'yes' or 'no'.")
        
Htype = input("what type of house are you using?")
Hcol = input("what color is it?")
Hroom = int(input("how many rooms does it have?"))
my_house = House(Htype, Hcol, Hroom)
my_house.introduce()
my_house.paint()
```
