```python
import customtkinter as ctk


class Delivery:
    def __init__(self, location, weight):
        self.location = location.strip().title()
        self.weight = weight

    def calculate_price(self):
        if self.location == "Pau":
            if self.weight >= 10:
                return 2000
            else:
                return 1500

        elif self.location == "Epe":
            if self.weight >= 10:
                return 5000
            else:
                return 4000

        else:
            return None


def check_price():
    location = location_entry.get()

    try:
        weight = float(weight_entry.get())
    except ValueError:
        result_label.configure(text="Please enter a valid number for weight.")
        return

    delivery = Delivery(location, weight)
    price = delivery.calculate_price()

    if price is None:
        result_label.configure(text="Sorry, we only deliver to PAU and Epe.")
    else:
        result_label.configure(
            text=f"Delivery to {delivery.location} with {delivery.weight}kg package costs N{price}."
        )


ctk.set_appearance_mode("dark")
ctk.set_default_color_theme("blue")

app = ctk.CTk()
app.title("Delivery Price Calculator")
app.geometry("600x420")

title_label = ctk.CTkLabel(
    app,
    text="Delivery Price Calculator",
    font=("Arial", 24, "bold")
)
title_label.pack(pady=20)

location_label = ctk.CTkLabel(
    app,
    text="Enter location (PAU or Epe):",
    font=("Arial", 16)
)
location_label.pack(pady=8)

location_entry = ctk.CTkEntry(
    app,
    width=300,
    height=40,
    placeholder_text="Enter location"
)
location_entry.pack(pady=8)

weight_label = ctk.CTkLabel(
    app,
    text="Enter package weight (kg):",
    font=("Arial", 16)
)
weight_label.pack(pady=8)

weight_entry = ctk.CTkEntry(
    app,
    width=300,
    height=40,
    placeholder_text="Enter weight"
)
weight_entry.pack(pady=8)

price_button = ctk.CTkButton(
    app,
    text="Calculate Price",
    command=check_price,
    width=200,
    height=40
)
price_button.pack(pady=15)

result_label = ctk.CTkLabel(
    app,
    text="",
    font=("Arial", 16),
    wraplength=500,
    justify="center"
)
result_label.pack(pady=20)

app.mainloop()
```
