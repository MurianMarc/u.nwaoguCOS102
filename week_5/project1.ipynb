```python
import customtkinter as ctk
import random


class Employee:
    def __init__(self, name):
        self.name = name.strip().title()
        self.employees = [
            "Mary Evans", "Eyo Ishan", "Durojaiye Dare", "Adams Ali",
            "Andrew Ugwu", "Stella Mankinde", "Jane Akibo", "Ago James",
            "Michell Taiwo", "Abraham Jones", "Nicole Anide", "Kosi Korso",
            "Adele Martins", "Emmanuel Ojo", "Ajayi Fatima"
        ]
        self.tasks = [
            "Loading",
            "Transporting",
            "Reviewing Orders",
            "Customer Service",
            "Delivering Items"
        ]

    def check_employee(self):
        return self.name in self.employees

    def take_attendance(self):
        return f"Attendance taken for {self.name}."

    def assign_task(self):
        return f"Task assigned: {random.choice(self.tasks)}"

    def refuse_access(self):
        return f"Sorry, {self.name}. You are not authorized to use this system."


def verify_employee():
    entered_name = name_entry.get()

    if entered_name.strip() == "":
        result_label.configure(text="Please enter a name.")
        return

    worker = Employee(entered_name)

    if worker.check_employee():
        attendance_message = worker.take_attendance()
        task_message = worker.assign_task()
        result_label.configure(
            text=f"Welcome, {worker.name}!\n{attendance_message}\n{task_message}"
        )
    else:
        result_label.configure(text=worker.refuse_access())


ctk.set_appearance_mode("dark")
ctk.set_default_color_theme("blue")

app = ctk.CTk()
app.title("Employee Verification System")
app.geometry("600x400")

title_label = ctk.CTkLabel(
    app,
    text="Employee Attendance and Task System",
    font=("Arial", 24, "bold")
)
title_label.pack(pady=20)

name_label = ctk.CTkLabel(
    app,
    text="Enter employee name:",
    font=("Arial", 16)
)
name_label.pack(pady=10)

name_entry = ctk.CTkEntry(
    app,
    width=300,
    height=40,
    placeholder_text="Type full name here"
)
name_entry.pack(pady=10)

verify_button = ctk.CTkButton(
    app,
    text="Verify Employee",
    command=verify_employee,
    width=200,
    height=40
)
verify_button.pack(pady=15)

result_label = ctk.CTkLabel(
    app,
    text="",
    font=("Arial", 16),
    wraplength=500,
    justify="center"
)
result_label.pack(pady=20)

app.mainloop()
```
