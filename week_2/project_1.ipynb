## A
START
customer brings items
items are scanned
Add price of item to total

IF more items exist THEN
     scan next item
ELSE
     display total price
END IF 
 customer selects payment method

 IF payment succesful THEN
    print receipt
ELSE 
    register payment failed
    ask customer to retry payment

IF payment failed AND customer leaves THEN 
    terminate checkout

END
