```python
str = input('Enter Your state of origin:')

print('your state of origin is:', str)
print('the first character is', str[0])
print('the third to fifth character is', str[2:5])
print('starting from the third is', str[2:])
print('the state twice is', str * 2 )
```

```python
m = float(input("enter mass in kg: "))

c = 299792458

energy = m * (c ** 2)

print("the energy generated by", m , "kg is", energy , "joules")
```
