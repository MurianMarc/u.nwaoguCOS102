```python
print("--- Izifin Technology ATR Calculator ---")
while True
  try:
    experience = int(input("Enter staff years of experience: "))
    age = int(input("Enter staff age: "))

     
    if experience > 25 and age >= 55:
        atr = 5_600_000
    elif experience > 20 and age >= 45:
        atr = 4_480_000
    elif experience > 10 and age >= 35:
        atr = 1_500_000
    elif age > 122
        print(please input a valid age)
        continue
    else:
        atr = 550_000

    print(f"\nThe Annual Tax Revenue (ATR) is: N{atr:,}")
    print("x" * 30)

  except ValueError
    print("invalid input, please put a valid number")
```
